Import dependencies

```python
import pandas as pd
from sklearn.externals import joblib
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

Import datasets

```python
X_train = pd.read_csv('./data/X_train.csv', index_col='index')
y_train = pd.read_csv('./data/y_train.csv', index_col='index')
X_test = pd.read_csv('./data/X_test.csv', index_col='index')
```

```python
regr = LinearRegression()
regr.fit(X_train, y_train)
```

```python
# report MSE for training set
```

```python
# coefficients
regr.coef_
```

Generate predictions using X_test dataset, and save to CSV using appropriate format.

```python
# Use trained model to generate predictions
y_pred = regr.predict(X_test)
```

```python
# intermediate step to create dataframe with named column and index column
y_pred = pd.DataFrame(y_pred, columns=['target'])
y_pred.head()
```

```python
# save to CSV in desired format
y_pred.to_csv('./data/y_pred.csv', index_label='index')
```

```python
# save trained model to file
joblib.dump(regr, 'RegressionModel.sav')
```
